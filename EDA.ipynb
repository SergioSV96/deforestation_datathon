```python
# Read image from df['example_path], show the df['label'] and the image
def show_image(df, index):
    # Read the image
    img = cv2.imread(df['example_path'][index])
    # Convert the image to RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # Show the image
    plt.imshow(img)
    # Show the label
    plt.title(df['label'][index])
    plt.show()

# Show a random image from the dataset
show_image(df, random.randint(0, len(df)))
```
